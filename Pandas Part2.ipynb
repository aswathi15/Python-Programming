```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv("C:\\Users\\Aswathi\\Desktop\\Movie-Ratings.csv")
```

```python
len(movies)
```

```python
movies.head()
```

```python
movies.columns = ['Film','Genre','CriticRating','AudienceRating','BudgetMillions','Year']
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.describe()
```

```python
movies.Film = movies.Film.astype('category')
movies.Genre = movies.Genre.astype('category')
movies.Year = movies.Year.astype('category')
```

```python
movies.info()
```

```python
movies.describe()
```

```python
#Joinplot
sns.jointplot(data=movies, x = 'CriticRating', y='AudienceRating',kind='hex',color ='red')
```

```python
#<<< Chart1 
```

```python
#Histogram
sns.distplot(movies['AudienceRating'],bins=15)
```

```python
n1 = plt.hist(movies['AudienceRating'],bins=15)
```

```python
n1 = plt.hist(movies['CriticRating'],bins=15)
```

```python
#Stacked Histogram
plt.hist(movies[movies['Genre'] == "Action"]['BudgetMillions'],bins=15)
plt.hist(movies[movies['Genre'] == "Drama"]['BudgetMillions'],bins=15)
plt.hist(movies[movies['Genre'] == "Thriller"]['BudgetMillions'],bins=15)

plt.show()
```

```python
genre_list = movies['Genre'].unique()
genre_list
```

```python
color_list = ['red','green','blue','yellow','pink','orange','black']
```

```python
for i,j in enumerate(genre_list) :
    plt.hist(movies[movies['Genre'] == j]['BudgetMillions'],bins=15,color = color_list[i],stacked=True)
plt.show()
    
```

```python
#KDE plot

sns.lmplot(x ='CriticRating',y ='AudienceRating',data=movies,hue='Genre',fit_reg=False,size=5,aspect=1,)
```

```python
sns.kdeplot(movies['CriticRating'],movies['AudienceRating'],shade=True,shade_lowest=False,cmap ='Reds')
```

```python
#ViolinPlots

sns.violinplot(x ='Genre',y='CriticRating',data=movies)
```

```python
sns.boxplot(x ='Genre',y='CriticRating',data=movies)
```

```python
#Creating a Facetgrid

g = sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
#g = g.map(plt.hist,'CriticRating','AudienceRating')
g = g.map(plt.hist,'BudgetMillions')
```

```python
g = sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
kws = dict(s=50,linewidth = 0.5, edgecolor = 'black')
g = g.map(plt.scatter,'CriticRating','AudienceRating',**kws)
g.set(xlim = (0,100), ylim=(0,100))
for ax in g.axes.flat:
    ax.plot((0,100),(0,100),c='black',ls ='--')
g.add_legend()
```
