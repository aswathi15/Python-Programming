```python
## Introduction to Seaborn
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
stats = pd.read_csv("C:\\Users\\Aswathi\\Desktop\\DemographicData.csv")
```

```python
stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers',\
       'IncomeGroup']
```

```python
#Distribution

vis1 = sns.distplot(stats['InternetUsers'],bins = 30)
```

```python
#Boxplots

vis2 = sns.boxplot(data=stats, x ='IncomeGroup',y='BirthRate')
```

```python
stats.head()
```

```python
vis3 = sns.lmplot(y = 'BirthRate',x = 'InternetUsers',data=stats,fit_reg=False,hue ='IncomeGroup',size = 5,scatter_kws ={"s":200})
```
